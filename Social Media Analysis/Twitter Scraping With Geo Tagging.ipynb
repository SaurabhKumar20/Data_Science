```python
import pandas as pd
import snscrape.modules.twitter as sntwitter
import itertools
```

```python
#Shortened code to fetch the first (i.e. most recent) 100 tweets that contains the phrase data science 
pd.DataFrame(itertools.islice(sntwitter.TwitterSearchScraper(
    '"data science"').get_items(), 100))
```

#### The TwitterSearchScraper uses Twitter‚Äôs advanced search; for an in depth run down of all its capabilities check out this table ‚Äî I cannot recommend it enough!

```python
#To find all tweets about pizza in Los Angeles. This can be achieved using the following code:

df_city = pd.DataFrame(itertools.islice(sntwitter.TwitterSearchScraper(
    'pizza near:"Los Angeles" within:10km').get_items(), 50))[['date', 'content']]
```

#### OR

```python
# Instead of using the city‚Äôs name you can use its coordinates:

loc = '34.052235, -118.243683, 10km'
df_coord = pd.DataFrame(itertools.islice(sntwitter.TwitterSearchScraper(
    'pizza geocode:"{}"'.format(loc)).get_items(), 50))[['date', 'content']]
```

```python
loc = '34.052235, -118.243683, 100km'
df_coord = pd.DataFrame(itertools.islice(sntwitter.TwitterSearchScraper(
    'GoDaddy geocode:"{}"'.format(loc)).get_items(), 5000))[['user', 'date','content']]

df_coord['user_location'] =  df_coord['user'].apply(lambda x: x['location'])
df_coord['user'] =  df_coord['user'].apply(lambda x: x['username'])
```

```python
df_coord
```

```python
df_coord.to_csv('tweets within 100 Km of Los Angeles location.csv')
```

```python
#Initally 'user' column consisted of dictionary of users in each row 
df_coord['user'][0]
```

```python
df= pd.DataFrame(itertools.islice(sntwitter.TwitterSearchScraper(
    'GoDaddy').get_items(), 10000))[['user','date', 'content']]

df['user_location'] =  df['user'].apply(lambda x: x['location'])
df
```

```python
df['user_location'].unique()
```

```python
df.to_csv('10000 tweets of GoDaddy with location.csv')
```
