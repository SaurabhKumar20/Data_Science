# Forecasting models, one by one:

## Naive Method of Forecasting

```python
dd= np.asarray(train['Number of Tractor Sold'])
y_hat = test.copy()
y_hat['naive'] = dd[len(dd)-1]
plt.figure(figsize=(12,8))
plt.plot(train.index, train['Number of Tractor Sold'], label='Train')
plt.plot(test.index,test['Number of Tractor Sold'], label='Test')
plt.plot(y_hat.index,y_hat['naive'], label='Naive Forecast')
plt.legend(loc='best')
plt.title("Naive Forecast")
plt.show()
```

#### Naive method forecast assumes that the next period forecast is the very last datapoint, and it continues like that.

```python
from sklearn import metrics
from math import sqrt
rmse=sqrt(metrics.mean_squared_error(test['Number of Tractor Sold'],y_hat['naive']))
print('RMSE for Naive Method is: ', rmse)
```

#### Observations:
    * Converting the data into Numpy array, so that the original test data do not get impacted.
    * y-hat: Creating a copy of the test data.
    * Column called naive is created in y-hat, with dd array's last value.
    * The green line above is the Naive Forecast.
    * RMSE value of the actuals vs forecast is being evaluated, and printed.
    * RMSE value of 155.3683794513328 is quite large. Also, the green line above is not quite close to the orange curve. So, its not a good forecast.

# Exponential Smoothing Method

```python
from statsmodels.tsa.api import ExponentialSmoothing, SimpleExpSmoothing, Holt
y_hat_avg = test.copy()
fit2 = SimpleExpSmoothing(np.asarray(train['Number of Tractor Sold'])).fit(smoothing_level=0.15,optimized=False)
y_hat_avg['SES'] = fit2.forecast(len(test)) # Simple Exponential Smoothing.
plt.figure(figsize=(16,8))
plt.plot(train['Number of Tractor Sold'], label='Train')
plt.plot(test['Number of Tractor Sold'], label='Test')
plt.plot(y_hat_avg['SES'], label='SES')
plt.legend(loc='best')
plt.show()
```

#### Simple Exponential smoothing assumes that there is no trend/no seasonality. Without those assumptions, it will try to fit into the data, and do the forecast.

```python
from sklearn import metrics
from math import sqrt
rmse=sqrt(metrics.mean_squared_error(y_hat['Number of Tractor Sold'],y_hat_avg['SES']))
print('RMSE for SES is: ', rmse)
```

#### Observations:
    * We can play with the smoothing level constant (alpha) to decrease the RMSE, and hence improve the model.
    * No trend or seasonality in data assumed in Exponential Smoothing.
    * Notice that the green forecast line is a straight line with zero slope, which means that there is no trend relationship in this forecast.

# Holt's Exponential Smoothing

```python
y_hat_avg = test.copy()

fit1 = Holt(np.asarray(train['Number of Tractor Sold'])).fit(smoothing_level = 0.1,smoothing_slope = 0.2)
y_hat_avg['Holt_linear'] = fit1.forecast(len(test))

plt.figure(figsize=(16,8))
plt.plot(train['Number of Tractor Sold'], label='Train')
plt.plot(test['Number of Tractor Sold'], label='Test')
plt.plot(y_hat_avg['Holt_linear'], label='Holt_linear')
plt.legend(loc='best')
plt.show()
```

```python
from sklearn import metrics
from math import sqrt
rmse=sqrt(metrics.mean_squared_error(y_hat['Number of Tractor Sold'],y_hat_avg['Holt_linear']))
print('RMSE for Holt_Linear_Expo_Smoothing is: ', rmse)
```

#### Observations:
    * Trend is assumed.
    * In Holt's, a tilted/slanting green line can be observed. This is the biggest difference between SES and Holt's.
      Holt assumes trend.
    * RMSE value reduced to 123.68.  

# Holt's Winters Method

```python
y_hat_avg = test.copy()
fit1 = ExponentialSmoothing(np.asarray(train['Number of Tractor Sold']) ,seasonal_periods=12 ,trend='add', seasonal='add',).fit(use_boxcox=True)
y_hat_avg['Holt_Winter'] = fit1.forecast(len(test))
plt.figure(figsize=(16,8))
plt.plot( train['Number of Tractor Sold'], label='Train')
plt.plot(test['Number of Tractor Sold'], label='Test')
plt.plot(y_hat_avg['Holt_Winter'], label='Holt_Winter')
plt.legend(loc='best')
plt.show()
```

## Observations:

    * Its able to capture the trend and seasonality in the data.
    * But it fails to capture offsets, irregularities in the data.

```python
from sklearn import metrics
from math import sqrt
rmse=sqrt(metrics.mean_squared_error(y_hat['Number of Tractor Sold'],y_hat_avg['Holt_Winter']))
print('RMSE for Holt_Linear_Expo_Smoothing is: ', rmse)
```

#### Observations:
    * The forecast in the test period fits the test data really well, with setting Seasonal_periods=12
    * Setting Seasonal_periods=12 means that we have assumed 12 seasons in a year.
    * Remarkable reduction in RMSE achieved by this Forecasting Method.

### STATIONARY TEST

```python
# Augmented Dickey Fuller Test
from statsmodels.tsa.stattools import adfuller
# from numpy import log
result=adfuller(df.dropna())
#result
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
```

#### With p-value>0.05, the null fails to get rejected. So, it is a non-stationary series.

## Building Auto ARIMA model

```python
arima_model=auto_arima(train, seasonal=False, 
                          error_action='ignore',trace = True,
                          supress_warnings=True)


# arima_model =  auto_arima(train,start_p=0, d=1, start_q=0, 
#                           max_p=5, max_d=5, max_q=5, start_P=0, 
#                           D=1, start_Q=0, max_P=5, max_D=5,
#                           max_Q=5, m=12, seasonal=True, 
#                           error_action='warn',trace = True,
#                           supress_warnings=True,stepwise = True,
#                           random_state=20,n_fits = 50 )

arima_model.fit(train)
```

```python
arima_model.summary()
```

```python
arima_model.order
```

```python
# pred = pd.DataFrame(arima_model.predict(start=df_new.index[0], end=df_new.index[-1]))

# pred
from statsmodels.tsa.arima_model import ARIMA

from tqdm import tqdm
order=arima_model.order
yhat=list()
for t in tqdm(range(len(test['Number of Tractor Sold']))):
    temp_train=df[:len(train)+t]
    model=ARIMA(temp_train['Number of Tractor Sold'], order=order)
    model_fit=model.fit(disp=False)
    predictions=model_fit.predict(start=len(temp_train), end=len(temp_train), dynamic=False)
    yhat=yhat+[predictions]
    
yhat=pd.concat(yhat)
# resultsDict['AutoARIMA {0}'.format(order)]=evaluate(test.pollution_today, yhat)

# predictionsDict['AutoARIMA {0}'.format(order)]=yhat.values
```

```python
plt.plot(test['Number of Tractor Sold'].values,label='Original')
plt.plot(yhat.values, color='red', label='AutoARIMA {0}'.format(order))
plt.legend()
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error

print("Mean Absolute Error:", mean_absolute_error(test,yhat))
print("RMSE:",np.sqrt(mean_squared_error(test,yhat)))
```

#### With Setting Seasonal component to False, Auto-ARIMA doesn't predict the model very well!

#### Now, lets go for AUTO-ARIMA with setting Seasonal component to True.

```python
df.plot(figsize=(10,5))
```

In the Auto ARIMA model, note that small p,d,q values represent non-seasonal components, and capital P, D, Q represent seasonal components. It works similarly like hyper tuning techniques to find the optimal value of p, d, and q with different combinations and the final values would be determined with the lower AIC, BIC parameters taking into consideration.
Here, we are trying with the p, d, q values ranging from 0 to 5 to get better optimal values from the model.

```python
arima_model =  auto_arima(train,start_p=0, d=1, start_q=0, 
                          max_p=5, max_d=5, max_q=5, start_P=0, 
                          D=1, start_Q=0, max_P=5, max_D=5,
                          max_Q=5, m=12, seasonal=True, 
                          error_action='warn',trace = True,
                          supress_warnings=True,stepwise = True,
                          random_state=20,n_fits = 50 )
```

```python
arima_model.summary()
```

## Forecasting on the test data

#### Using the trained model which was built in the earlier step to forecast the sales on the test data.

```python
prediction = pd.DataFrame(arima_model.predict(n_periods = 12),index=test.index)
prediction.columns = ['predicted_sales']
prediction
```

```python
plt.figure(figsize=(8,5))
plt.plot(train,label="Training")
plt.plot(test,label="Test")
plt.plot(prediction,label="Predicted")
plt.legend(loc = 'Left corner')
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error

print("Mean Absolute Error:", mean_absolute_error(test,prediction))
print("RMSE:",np.sqrt(mean_squared_error(test,prediction)))
```

## SARIMAX model Implementation
