#### The argument fit=norm tells us whether this is close to a normal distribution or not!  Although we don't like our variables to be normal or not. What we would like to be normal is the error terms.

```python
X = dataset[['Petrol_tax','Paved_Highways','Average_income','Population_Driver_licence(%)']]#'Petrol_tax','Paved_Highways'
y = dataset['Petrol_Consumption']
```

#### These are the four independent, and 1 dependent variable.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state=456)
```

#### Observations: 
    This part does the splitting. Test-size=0.2 means that 20% of the data for testing, and random_state=456, is like a 
    sample seed (in R) Lets say we don't have random state, and we run this. Lets say rows 1,3,5,30... are selected in test. When we run that again different set of rows will be obtained. Different set of rows will be obtained. Its a random   
    step, its happening again, a different set of train and test will be obtained, which means that everytime it is run, 
    there is going to be a slight change in the formulation of what's getting inside the train and what's going inside test. As a result of which training of model will change.
    As a result, your training of your model will change slightly because you are training your model with different data, and your model results will fluctuate, which means you don't want to come across a scenario, where same model is showing different accuracies at two different times. You ran it again and couldn't replicate the results. Though its a random 
    step, these kind of fluctuations can happen. If these fluctuations are very high, thats something one needs to be wary
    of. Somewhere you have not ensured that your assumptions are right. You are getting such high fluctuations for some
    random steps. That's not good sign. But in case we want to ensure replicablility, this can be done.
    random_state= can be any number. So, the next time you run it with the same number, you will get the same split. This is the purpose of random state.
     
    -> X_train, X_test, y_train, y_test 
    These four dataframes will first build the model.
    

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression(fit_intercept = True)
regressor.fit(X_train, y_train)
```

#### Observations:
    ->First fitting the model in the regressor. They'll be fitted with the x-train and the y-train.
    ->When the model is going to predict, x_test is passed to it.
    ->It has taken 80% of x and y.
    ->For now, we are passing the x_train and the y_train, so that the model can learn based on 80% of the data.

#### First predicting using SK Learn. We are running Linear Regression function using SK Learn. This SK Learn is using SGD(Scholastic Gradient Descent) through its estimation. 

```python
coeff_df = pd.DataFrame(regressor.coef_, X.columns, columns=['Coefficient'])
coeff_df
```

```python
regressor.intercept_
```

```python
y_pred = regressor.predict(X_train)
df = pd.DataFrame({'Actual': y_train, 'Predicted': y_pred})
df
```

```python
# Validating OLS Assumptions
plt.scatter(y_pred, (y_train-y_pred))
plt.xlabel("Fitted values")
plt.ylabel("Residuals")
```

```python
(y_train-y_pred).mean()
```

```python
#Residual plot check is imp. to verify the OLS assumptions.
sns.distplot(y_train-y_pred , fit=norm);
plt.xlabel('Residuals')
```

#### Normal curve is drawn in black, fit curve is drawn in blue

#### This residual plot is very close to normal one.

```python
from scipy import stats
stats.probplot(y_train-y_pred, plot=plt)
plt.show()
```

#### This plot tells whether the residual curve is normal or not. 

```python
import statsmodels.api as sm
```

```python
# Unlike sklearn that adds an intercept to our data for the best fit, statsmodel doesn't. We need to add it ourselves
# Remember, we want to predict the price based off our features.
# X represents our predictor variables, and y our predicted variable.
# We need now to add manually the intercepts
X_endog = sm.add_constant(X_train)
X_endog1 = sm.add_constant(X_test)
```

```python
res = sm.OLS(y_train, X_endog)
res.fit()
```

#### Linear model is being fit here, with the help of OLS function.

```python
#Final model summary obtained
res.fit().summary()
```

```python
# Build a model keeping in mind the individual t tests
```

```python
# Model evaluation metrics - Regression problems

y_pred = regressor.predict(X_test)

from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
def mean_absolute_percentage_error(y_true,y_pred):
    return    np.mean(np.abs((y_true-y_pred)/y_true))*100
```

```python
print('MAPE:',mean_absolute_percentage_error(y_test,y_pred))
```

```python
# Model evaluation metrics - Regression problems
# Notice test -> train

y_pred = regressor.predict(X_train)

from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_train, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_train, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_train, y_pred)))
```

```python
# Calculating MAPE for y_train
print('MAPE:',mean_absolute_percentage_error(y_train,y_pred))
```
