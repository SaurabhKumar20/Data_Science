# Concate and Merge

```python
import numpy as np
import pandas as pd
```

```python
students = pd.DataFrame({'id_number' : [204, 202, 264, 136, 238],
                               'Name' : ['James Smith',
                                        'Maria Rodriguez',
                                        'David Smith', 
                                        'Patricia Garcia',
                                        'Jessica Hernandez'],
                               'Age' : [26, 22 ,24, 21, 25],
                            'Gender' : ['M', 'F', 'M', 'F', 'F'],
                    
                        })
```

```python
students
```

```python
professors = pd.DataFrame({'Prof_name' : ['Thomas Smith',
                                          'Margaret Garcia',
                                          'James Johnson', 
                                          'Richard Miller',
                                          'Margaret MOORE'],
                      'id_number' : [204, 264, 108, 238, 136],
                     'Department' : ['ECE', 'EE' ,'CSE', 'ME', 'CE'],
                      
                         })
```

```python
professors
```

```python
pd.concat([students, professors], sort = True)
```

```python
pd.concat([students, professors], sort = True, ignore_index = True)
```

```python
pd.merge(students, professors)
```

```python
pd.merge(students, professors, how = 'left')
```

```python
pd.merge(students, professors, how = 'right')
```

```python
pd.merge(students, professors, how = 'inner')
```

```python
pd.merge(students, professors, how = 'outer')
```

# Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
arr1 = np.arange(5, 15)
arr1
```

```python
arr2 = arr1 ** 2
arr2
```

```python
# Creating the the figure/canvas
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.9, 0.9])   # left, bottom, width, height (range 0 to 1)

axes.plot(arr1, arr2)
axes.set_xlabel('Arr1')
axes.set_ylabel('Arr2')
axes.set_title('Arr1 vs Arr2')
```

```python
# Creating the the figure/canvas
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.9, 0.9])   # left, bottom, width, height (range 0 to 1)
axes2 = fig.add_axes([0.2, 0.5, 0.3, 0.4]) 

axes1.plot(arr1, arr2)
axes1.set_xlabel('Arr1')
axes1.set_ylabel('Arr2')
axes1.set_title('Arr1 vs Arr2')

axes2.plot(arr2, arr1)
axes2.set_xlabel('Arr2')
axes2.set_ylabel('Arr1')
axes2.set_title('Arr2 vs Arr1')
```

```python
# Creating the the figure/canvas
fig, axes = plt.subplots(nrows = 2, ncols = 3)

axes[0][0].plot(arr1, arr2)
axes[0][0].set_xlabel('Arr1')
axes[0][0].set_ylabel('Arr2')
axes[0][0].set_title('Arr1 vs Arr2')

axes[1][2].plot(arr2, arr1)
axes[1][2].set_xlabel('Arr2')
axes[1][2].set_ylabel('Arr1')
axes[1][2].set_title('Arr2 vs Arr1')

plt.tight_layout()
```

```python
# Creating the the figure/canvas
#  Controlling the figure size with figsize attribute
fig, axes = plt.subplots(nrows = 2, ncols = 3, figsize = (10,8),dpi = 100)

axes[0][0].plot(arr1, arr2)
axes[0][0].set_xlabel('Arr1')
axes[0][0].set_ylabel('Arr2')
axes[0][0].set_title('Arr1 vs Arr2')

axes[1][2].plot(arr2, arr1)
axes[1][2].set_xlabel('Arr2')
axes[1][2].set_ylabel('Arr1')
axes[1][2].set_title('Arr2 vs Arr1')

plt.tight_layout()
```

# legends

```python
# Creating the the figure/canvas
fig, axes = plt.subplots(figsize = (6,5))

axes.plot(arr1, arr2, label = 'Arr1 vs Arr2')
axes.set_xlabel('Arr1')
axes.set_ylabel('Arr2')
axes.set_title('Arr1 vs Arr2')

axes.plot(arr1, arr1, label = 'Arr1 vs Arr1')
axes.set_xlabel('Arr1')
axes.set_ylabel('Arr1')
axes.set_title('Arr1 vs Arr1')

axes.legend(loc = 0)
```
